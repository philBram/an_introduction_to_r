```r
df <- mtcars
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)
par(mfcol = c(2, 1), cex.main = 2, cex.lab = 1.5, cex.axis = 1.5)

hist(df$wt, xlab = 'Wt', main = 'Histogram of Wt')
grid(col = 'black')
hist(df$mpg, xlab = 'mpg', main = 'Histogram of mpg')
grid(col = 'black')
```

```r
options(repr.plot.width = 15, repr.plot.height = 8)

plot(df$wt, df$mpg, cex.lab = 1.5, cex.axis = 1.5)
```

```r
plot(df$wt, df$mpg, 
     cex.lab = 1.5, 
     cex.axis = 1.5, 
     cex.main = 2, 
     main = 'MPG as a Function of Weight of Cars', 
     xlab = 'Weight (in 1000 pounds)', 
     ylab = 'MPG', 
     pch = 19, 
     lwd = 10, 
     col = '#cc0000')
```

```r
scatter.smooth(df$wt, df$mpg,
              col = '#cc0000',
              pch = 19,
              lwd = 8,
              main = 'MPG as a Function of Weight of Cars',
              xlab = 'Weight (in 1000 pounds)',
              ylab = 'MPG',
              cex.main = 2,
              cex.lab = 1.5,
              cex.axis = 1.5)
```
